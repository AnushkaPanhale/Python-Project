# Performing Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of courses across subjects
plt.figure(figsize=(10,6))
sns.countplot(y='subject', data=data)
plt.title('Distribution of Courses by Subject')
plt.show()

# Distribution of courses across levels (if available)
plt.figure(figsize=(10,6))
sns.countplot(y='level', data=data)
plt.title('Distribution of Courses by Level')
plt.show()
```

 # Exploring Course Pricing Strategies

Analyze the pricing strategy based on subjects:

```python
# Distribution of prices
plt.figure(figsize=(10,6))
sns.histplot(data['price'], bins=30, kde=True)
plt.title('Price Distribution')
plt.show()

# Distribution of number of subscribers
plt.figure(figsize=(10,6))
sns.histplot(data['num_subscribers'], bins=30, kde=True)
plt.title('Number of Subscribers Distribution')
plt.show()

# Relationship between number of subscribers and reviews
plt.figure(figsize=(10,6))
sns.scatterplot(x='num_subscribers', y='num_reviews', data=data)
plt.title('Subscribers vs. Reviews')
plt.show()
```

# Subscriber Engagement



Explore which subjects have the highest subscriber engagement:

```python
# Average number of subscribers per subject
avg_subscribers_per_subject = data.groupby('subject')['num_subscribers'].mean().sort_values()
plt.figure(figsize=(10,6))
avg_subscribers_per_subject.plot(kind='barh')
plt.title('Average Subscribers per Subject')
plt.xlabel('Average Subscribers')
plt.show()
```

# Correlations Among Numerical Variables

 Check the correlation between numerical variables like num_subscribers, num_reviews, and content_duration:

```python
numerical_features = data.select_dtypes(include=['int64', 'float64']).columns
```

```python
# Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(data[numerical_features].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

#  Analyzing Free vs. Paid Courses


Check how many courses are free versus paid and explore their characteristics:

```python
# Count of free vs. paid courses
plt.figure(figsize=(6,6))
sns.countplot(x='is_paid', data=data)
plt.title('Free vs. Paid Courses')
plt.xlabel('Is Paid')
plt.ylabel('Count')
plt.show()

# Average subscribers for free vs. paid courses
avg_subscribers_paid = data.groupby('is_paid')['num_subscribers'].mean()
avg_subscribers_paid.plot(kind='bar', color=['blue', 'green'], figsize=(8,6))
plt.title('Average Subscribers: Free vs. Paid Courses')
plt.xlabel('Is Paid')
plt.ylabel('Average Number of Subscribers')
plt.show()
```

# Course Duration Analysis

Examine the distribution of course durations and their relation to subscriber count:

```python
# Distribution of content duration
plt.figure(figsize=(10,6))
sns.histplot(data['content_duration'], bins=30, kde=True)
plt.title('Content Duration Distribution')
plt.xlabel('Content Duration (hours)')
plt.show()

# Scatter plot: Content duration vs. number of subscribers
plt.figure(figsize=(10,6))
sns.scatterplot(x='content_duration', y='num_subscribers', data=data, hue='subject')
plt.title('Content Duration vs. Number of Subscribers')
plt.xlabel('Content Duration (hours)')
plt.ylabel('Number of Subscribers')
plt.legend(title='Subject', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Analyzing Course Reviews

Explore the relationship between reviews and subscribers:

```python
# Scatter plot: Number of subscribers vs. number of reviews
plt.figure(figsize=(10,6))
sns.scatterplot(x='num_reviews', y='num_subscribers', data=data, hue='is_paid')
plt.title('Number of Reviews vs. Number of Subscribers')
plt.xlabel('Number of Reviews')
plt.ylabel('Number of Subscribers')
plt.legend(title='Is Paid', loc='upper left')
plt.show()
```

# Analysis by Course Level

Calculate total revenue generated by courses in each subject (if prices are in the dataset):

```python
# Count of courses by level
plt.figure(figsize=(8,6))
sns.countplot(y='level', data=data, order=data['level'].value_counts().index)
plt.title('Distribution of Courses by Level')
plt.xlabel('Count')
plt.show()

# Average subscribers per level
avg_subscribers_per_level = data.groupby('level')['num_subscribers'].mean().sort_values()
plt.figure(figsize=(10,6))
avg_subscribers_per_level.plot(kind='barh', color='purple')
plt.title('Average Subscribers per Course Level')
plt.xlabel('Average Subscribers')
plt.show()
```
